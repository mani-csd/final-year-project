```python
import pandas as pd

df = pd.read_csv(r"D:\marketing_campaign_project\data\raw\marketing_and_product_performance.csv")

```

Create Target

```python
df["Conversion_Rate"] = df["Conversions"] / (df["Clicks"] + 1)

df["Customer_Response"] = ((df["Conversion_Rate"] > 0.05) & (df["ROI"] > 1)).astype(int)
```

Drop ID columns

```python
id_cols = [c for c in df.columns if "ID" in c]
df = df.drop(columns=id_cols)
```

Handle missing

```python
df = df.fillna(0)
```

```python
df.to_csv("D:/marketing_campaign_project/data/processed/cleaned_data.csv")
```
